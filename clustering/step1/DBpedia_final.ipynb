```python
ds=pd.read_csv('s3://vivek2511/lat_longs.txt',delimiter='[\t ]',engine='python',header=None)
```

```python
ds.head()
```

```python
ds.columns=['latitude','longitude','url']
```

```python
ds
```

```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
fig=plt.figure(figsize=(12,9))

m=Basemap(projection='mill',llcrnrlat=-90,urcrnrlat=90,llcrnrlon=-180,urcrnrlon=180,resolution='c')
m.drawcoastlines()
m.drawparallels(np.arange(-90,90,10),labels=[True,False,False,False])
m.drawmeridians(np.arange(-180,180,30),labels=[0,0,0,1])
m.scatter(ds['longitude'].tolist(),ds['latitude'].tolist(),latlon=True)
plt.title('Basemap tutorial',fontsize=20)
plt.show()
```
